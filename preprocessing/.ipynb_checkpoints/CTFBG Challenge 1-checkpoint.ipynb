```python
import xml.etree.ElementTree as ET
import re

def process_article(filename):
    
    tree = ET.parse(filename)
    text = tree.getroot()[1][3][7].text
    
    string = ""
    for c in text:
        string += c
    
    string = re.sub(r'\n', '', string)
    string = re.sub(r'\{\{.*?\}\}', '', string)
    string = re.sub(r'\<.*?\>', '', string)
    string = re.sub(r'\[\[', '', string)
    string = re.sub(r'\]\]', '', string)
    string = re.sub(r'\|', ' ', string)
    string = string.lower()
    print(string)
    
```

```python
process_article('articles/Cat')
```
